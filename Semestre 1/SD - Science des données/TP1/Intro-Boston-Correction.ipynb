## Apprendre un modèle linéaire

```python
from sklearn import linear_model
```

On construit un objet qui pourra apprendre un modèle de régression linéaire à partir de données.

```python
lr = linear_model.LinearRegression()
```

On s'intéresse à de la **prédiction**. Donc il est important de savoir comment se comportera le modèle quand il recevra de **nouvelles données**. Nous n'avons qu'un échantillon. Nous allons donc simuler l'apparition de nouvelles données en découpant l'échantillon en deux parties : 

- une partie qui sert à apprendre le modèle, l'ensemble d'apprentissage (train)
- une partie qui sert à estimer l'erreur sur de nouvelles données, l'ensemble test (test)

```python
from sklearn.model_selection import train_test_split
```

Le découpage est aléatoire, pour la reproductibilité on utilise un graine fixée pour le générateur de nombres aléatoires.

```python
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.33, random_state=42)
```

On apprend le modèle avec l'ensemble d'apprentissage.

```python
lr.fit(X_train, y_train)
```

```python
lr.intercept_, lr.coef_
```

On peut utiliser ce modèle pour prédire une valeur

```python
predictions_train = lr.predict(X_train)
```

```python
predictions_train[0]
```

```python
y_train[0]
```

On estime quelle est l'erreur sur cet ensemble train, puis sur l'ensemble de test

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
predictions_test = lr.predict(X_test)
```

```python
((predictions_test - y_test)**2).sum()
```

```python
((y_test - y_test.mean())**2).sum()
```

```python
1-((predictions_test - y_test)**2).sum() / ((y_test - y_test.mean())**2).sum()
```

On peut tracer un graphique montrant de combien on s'écarte de

```python
plt.scatter(predictions_test,y_test)
```

```python
fig, ax = plt.subplots()  # Plusieurs figures en une
ax.scatter(y_test, predictions_test)  # les points sur la diagonales sont exactement prédits
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])  # tracer la diagonale
ax.set_xlabel('Test')
ax.set_ylabel('Prediction')
plt.show()
```

On peut répéter cela 10 fois par validation croisée...

```python
from sklearn.model_selection import cross_val_predict
```

```python
y = boston.target

# cross_val_predict returns an array of the same size as `y` where each entry
# is a prediction obtained by cross validation:
predicted = cross_val_predict(lr, boston.data, y, cv=10)

fig, ax = plt.subplots()
ax.scatter(y, predicted, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()])
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
